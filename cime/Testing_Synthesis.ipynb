```python
from sympy import *
from sympy.geometry import *
from sympy.plotting.plot import Plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_segments(*segs, **kwargs):
    for seg in segs:
        x1,y1 = seg.p1
        x2,y2 = seg.p2
        plt.plot([x1,x2],[y1,y2], **kwargs)
```

```python
coupler_pos1 = Segment((0,100), (100,100))
coupler_pos2 = coupler_pos1.translate(200,-200)
coupler_pos2 = coupler_pos2.rotate(pi/6, coupler_pos2.p1)
segi = Segment(coupler_pos1.p1, coupler_pos2.p1)
segj = Segment(coupler_pos1.p2, coupler_pos2.p2)

iperp = segi.perpendicular_line(segi.midpoint)
jperp = segj.perpendicular_line(segj.midpoint)

pini = iperp.p2
pinj = jperp.p2

link2_pos1 = Segment(coupler_pos1.p1, pini)
link4_pos1 = Segment(coupler_pos1.p2, pinj)

link2_pos2 = Segment(coupler_pos2.p1, pini)
link4_pos2 = Segment(coupler_pos2.p2, pinj)
```

```python
plot_segments(link2_pos1, link4_pos1, coupler_pos1, color="r")
plot_segments(link2_pos2, link4_pos2, coupler_pos2, color="b")
plot_segments(iperp, jperp, segi, segj, ls="--", color="#aaaaaa")
plt.axis('equal')
```
